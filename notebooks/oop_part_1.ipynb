```python
#Class for an ATM machine

#Pascal Case for naming the class. In pascal case we start the word with capital letter.
#In class, only two things can come- 1.Data 2. Functions

#Non-Parameterized means it is not accepting any paramters
class Atm:
    def __init__(self):    #__init__ is a constructor (special function) which runs automatically without being called unlike a normal function 
        #print(id(self))
        self.pin=""
        self.balance=0
        self.menu()

    def menu(self):
        user_input=int(input("""
        Hello, Please enter your choice!
        Press 1 to create a pin
        Press 2 to check the balance
        Press 3 to deposite the amount
        Press 4 to withdraw the amount
        Press 5 to exit
        """))

        if user_input==1:
            self.create_pin()
        elif user_input==2:
            self.check_balance()
        elif user_input==3:
            self.deposite()
        elif user_input==4:
            self.withdrawal()
        elif user_input==5:
            exit
            
    def create_pin(self):
        user_pin=input("Please create your pin number: ")
        self.pin=user_pin
        print("Pin created successfully!")
        self.menu()
    def check_balance(self):
        user_pin=input("Please confirm your pin first: ")
        if user_pin==self.pin:
            print("Your current balance is: ",self.balance)
        else:
            print("Wrong Pin!!!")
        self.menu()
    def deposite(self):
        user_pin=input("Please confirm your pin first: ")
        if user_pin==self.pin:
            amount=int(input("Enter the amount you want to deposite: "))
            self.balance=amount+self.balance
            print("Amount added successfully")
        else:
            print("Wrong Pin!!!")
        self.menu()
    def withdrawal(self):
        user_pin=input("Please confirm your pin first: ")
        if user_pin==self.pin:
            withdrawal_amount=int(input("Enter the amount you want to withdraw: "))
            if self.balance>=withdrawal_amount:
                self.balance=self.balance-withdrawal_amount
                print("Your amount has been deducted from the main balance. Remaining balance: ",self.balance)
            else:
                print("Not enough money")
        else:
            print("Wrong Pin!!!")
        self.menu()
```

```python
#Fraction Numbers

#Parameterized constructor meaning it has parameters.
class Fraction:
    def __init__(self,x,y):
        self.num=x
        self.deno=y

    def __str__(self):    #The __str__ magic method automatically returns whatever we have inside the return, whenever an object is inside a print function
        return "{}/{}".format(self.num,self.deno)
        
    def __add__(self,other): #It means one is the self object which has it own num,deno and other is another object that has its own num, deno.
        new_num=self.num*other.deno+self.deno*other.num
        new_deno=self.deno*other.deno
        return "{0}/{1}".format(new_num,new_deno)

    def __sub__(self,other):
        new_num=self.num*other.deno-self.deno*other.num
        new_deno=self.deno*other.deno
        return "{0}/{1}".format(new_num,new_deno)

    def __mul__(self,other):
        new_num=self.num*other.num
        new_deno=self.deno*other.deno
        return "{0}/{1}".format(new_num,new_deno)

    def __truediv__(self,other):
        new_num=self.num*other.deno
        new_deno=self.deno*other.num
        return "{0}/{1}".format(new_num,new_deno)
```

```python
f1=Fraction(3,4)
f2=Fraction(5,6)
print(f1+f2)
```

```python
print(f1-f2)
print(f1*f2)
print(f1/f2)
```
