```python
#Write OOP classes to handle the following scenarios:

#· A user can create and view 2D coordinates
#· A user can find out the distance between 2 coordinates
#· A user can find find the distance of a coordinate from origin
#· A user can check if a point lies on a given line
#· A user can find the distance between a given 2D point and a given line


import math
class Coordinates:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __str__(self):
        return "({x},{y})".format(x=self.x,y=self.y)

    def distance(self,other):
        d=math.sqrt((other.x-self.x)**2 + (other.y-self.y)**2)
        return d

    def distance_origin(self):
        d=math.sqrt((self.x)**2 + (self.y)**2)
        return d


class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return "{}x + {}y + {}".format(self.a,self.b,self.c)

    def point_on_line(self,point:Coordinates):   #To use one class as an argument in another class
        #self.Coordinates(x,y)
        if self.a*point.x+self.b*point.y+self.c==0:
            return True
        else:
            return False

    def distance_line_point(self,point:Coordinates):
        nume=abs(self.a*point.x+self.b*point.y+self.c)
        deno=math.sqrt((self.a)**2+(self.b)**2)
        return nume/deno
```

```python
L=Line(1,1,-2)
print(L)
```

```python
C=Coordinates(1,10)
print(C)
```

```python
L.point_on_line(C)
```

```python
L.distance_line_point(C)
```

```python
#Pass by reference
class greet:
    def __init__(self,name):
        self.name=name

#It is a function not a method as it is outside the class
def greetings(x):
    print("Hello, {}".format(x.name))
    x=greet("Nikki")
    return x

a=greet("Mukesh")  #I created an object of greet class and store it into an reference varible a
x=greetings(a)   #Now, I just pass the reference variable to function and also x=greetings(a) means the return
#value will be stored in x.

print(x.name)
```

```python
#Encapsulation means protecting the data by controlling the access through methods.

class hey:
    def __init__(self,name):
        self.__name=name   #With the help of this __ syntax we can hide name instance varibale from the obj. 
        self.greetings()
        
    def greetings(self):
        print("Hello, {}".format(self.__name))
        
```

```python
z=hey("Don")
```

```python
#Now, even if someone use z.__name, it won't affect the code becuase when we use 
#__instance, python replace the instance varibale with _Classname__Instance.

z._hey__name="Jake" #By using this we can change the value of name again.
z.__name="Paul"    #It is creating a new instance variable, not changing the original one.
z.greetings()
```

```python
#Getter and Setter (If we want to give the access of name changing)

class hey_man:
    def __init__(self,name):
        self.__name=name   #With the help of this __ syntax we can hide name instance varibale from the obj. 
        self.greetings()

    def get(self):
        return self.__name
        
    def set(self,name):
        if type(name)==str:     #Other method if isinstance(name,str):
            self.__name=name
        else:
            return "Noep! Not allowed"
    
    def greetings(self):
        print("Hello, {}".format(self.__name))
```

```python
#Static Variable vs Instance Variable

#Objects can be stored in Lists and Dictionaries

class Person_Guide:
    #Static Varible
    __serial_num=0
    def __init__(self,name,gender):
        self.name=name
        self.gender=gender
        self.__serial_num=Person_Guide.__serial_num
        Person_Guide.__serial_num=Person_Guide.__serial_num+1

    def get():
        return Person_Guide.__serial_num
    def set(new_num):
        Person_Guide.__serial_num=new_num
```

```python
p1=Person_Guide("Giyan","Male")
p2=Person_Guide("Shizuka","Female")
p3=Person_Guide("Nobita","Male")
p4=Person_Guide("Sunio","Male")
p5=Person_Guide("Doremon","Raccoon")
```
