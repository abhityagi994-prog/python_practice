```python
#1. Maximum sum subarray
#Means we have to find out the subarray whose sum is maximum. Also, subarrays means the
#continuous numbers in a ordered sequence, not the random numbers.
L = [-2,4,7,-1,6,-11,14, 3, -1, -6]
d={}
for i in range(0,len(L)):
    subarray=[]
    for j in range(i,len(L)):
        subarray.append(L[j])
        d[sum(subarray)]=subarray
print(d)
print("\n")
highest=max(d)  #As we know in dictionaries everything is about keys. 
#And our keys overe here are sum of subarrays

for i,j in d.items():
    if i==highest:
        print(j,":",i)


#The order of Growth of this is O(n)*O(n)= O(n^2)
```

```python
#Kadane's Algorithm

L = [-2,4,7,-1,6,-11,14, 3, -1, -6]
curr_sum=0
curr_set=[]
best_sum=L[0]
best_set=[]

for i in L:
    if i+curr_sum>i:
        curr_sum=curr_sum+i
        curr_set.append(i)
    else:
        curr_sum=i
        curr_set.clear()
        curr_set.append(i)

    #print(curr_set,curr_sum) 
    
    if curr_sum>best_sum:
        best_sum=curr_sum
        best_set=curr_set

print(best_set,best_sum)

#The order of growth for this is O(n)
```

```python
# 2. Find element with left side smaller/right side greater in an array

L = [3,1,2,5,8,7,9] 

#Skipping the first ans last element becuase there is no left and right element to compare for them
for i in range(1,len(L)-1):  
    flag=True
    for j in range(0,i):
        if L[j]>L[i]:
            flag=False
    for k in range(i+1,len(L)):
        if L[k]<L[i]:
            flag=False

    if flag:
        print(L[i])

#Also, the time complexity for this code is O(n^2)
```

```python
#More pythonic way
L = [3,1,2,5,8,7,9] 
for i in range(1,len(L)-1):
    if max(L[:i]) < L[i] < min(L[i+1:]):
            print(L[i])


#Time complexity is still O(n^2), because we have min and max, which will run throw the entire loop to look
#for the min and max. First O(n) for max and second O(n) for min. So, it according to the law of addition, it
#would be 2n. And for loop it would be O(n). Hence, answer would be O(n^2).
```

```python
#Solution for order of n. O(n)
L = [3,1,2,5,8,7,9] 

max_list=[]
min_list=[]

max_num=L[0]
min_num=L[-1]


for i in L:
    if i>max_num:
        max_num=i
    max_list.append(max_num)
print(max_list)

for j in range(len(L)-1,-1,-1):
    if L[j] <min_num:
        min_num=L[j]
    min_list.insert(0,min_num)
print(min_list)

for i in range(1,len(L)-1):
    if max_list[i-1]<L[i]<min_list[i+1]:
        print(L[i])

#There are three different loops, but not loop inside loop. So, Big O would be O(n)+O(n)+O(n)=2n=n
```

```python
# 6. Rotate array to left d items

L = [1,2,3,4,5]
rotate = 2

for i in range(0,rotate):
    temp=L[0]
    for j in range(0,len(L)-1):
        L[j]=L[j+1]
    L[len(L)-1]=temp

print(L)
```
