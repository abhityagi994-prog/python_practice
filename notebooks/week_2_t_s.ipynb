```python
#Tuples
#A tuple in Python is similar to a list. The difference between the two is that we cannot change the 
#elements of a tuple once it is assigned whereas we can change the elements of a list.
#In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

#Characterstics

#· Ordered
#· Unchangeble
#· Allows duplicate
```

```python
#Creating a tuple
t=()
print(type(t))
```

```python
#Creating a single element tuple
t=(2)
print(type(t))

t=(2,)   #We have to do , in order to make it a tuple
print(type(t))
```

```python
#Homogenous Tuple 
t=(1,5,2,9,7,2,1)
print(t)
#Hetrogenous Tuple
t1=(2,"Hello",47,23.5,True)
print(t1)
```

```python
#2D tuple
t=(1,8,2,(5,7))
print(t)
#3D tuple
t=(1,8,2,(4.7,("OpenAI","Ice"),True))
print(t)
```

```python
#Type conversion
L=["Hello moto","Rocket"]
print(type(L))
a=tuple(L)
print(a)
print(type(a))
```

```python
#Accessing the items in tuples
#Indexing

t=(1,8,2,(4.7,("OpenAI","Ice"),True))
#Printing the OpenAI
print(t[-1][-2][0])

#Slicing
print(t[1:3])
print(t[-1::-1])
print(t[1:])
```

```python
#Editing the items in tuples

t=(1,5,3,8,1)
#Now, if we do like this 
t[2]=100
print(t)
#It will throw an error. Hence, it means tuples are immutable, just like strings
```

```python
#We cannot add the items as well in the tuples. As again they are immutable
```

```python
#Deleting 
t_=(2,3,7,"Dream_11")
del(t_)
print(t_)
#When we will try to print t_, it will show an error that t_ is not defined, which means t_ has been
#successfully deleted.
```

```python
#Operations in Tuples
#Arithmetic Operation

t1=(1,2,3)
t2=(4,5,6)
print(t1+t2)
#Addition works in tuples. While adding two tuples, the result is a new tuple. Because as we know tuples are 
#immutable. 

#Multiplication Opeartion
print(t1*3)
```

```python
#Loops in tuples

t1=(1,2,3)
for i in t1:
    print(i)
```

```python
t1=(1,2,3)
for i in t1[::-1]:
    print(i)
```

```python
#Comparison Operation

t1=(1,2,3)
t2=(4,5,6)
t=()
print(t1>t2)
print(t1<t2)
print(t1==t2)
print(t1!=t2)
```

```python
#Logical Operators
t1=(1,2,3)
t2=(4,5,6)
t=()
print(t1 and t2)
print(t1 and t)
print(t1 or t)
print(not t)
#Pyhton treats the empty tuple as False
```

```python
#Membership Operators
t1=(1,2,3)
t2=(4,5,6)
t=(1,8,2,(4.7,("OpenAI","Ice"),True))

print(1 in t1)
print(5 in t2)
print(4.7 not in t)
print(8 not in t)
```

```python
#Tuples function
#min/max, len, count, sort, sum
t=(1,8,2,4)
print(max(t))
print(min(t))

#Sum
print(sum(t))
```

```python
#Count
t=(5,1,9,3,6,2,2,1)
t.count(9)
```

```python
#For finding the index value of an element
print(t.index(1))
```

```python
#Sorted 
t=(7,2,9,8,10,1,4,4)
print(sorted(t,reverse=True))  #Sorted always convert the output into list format
```

```python
#Difference between Lists and Tuples

#· Syntax - Same
#· Mutability - Lists are mutable and tuples are immutable
#· Speed - Tuples are faster as compared to lists
#· Memory - Tuples occupr less memory as compared to tuples
#. Built in functionality - Lists has a lot of build-in-functions
#· Error prone - List are more prone to errors as compared to tuples
#· Usability 
```

```python
#Speed - When its comes to speed tuples are faster as they are immutable datatypes
#Memory Consumption - Tuples occupy less space as compared to Lists.


import sys

L=list(range(1,10000))
t=tuple(range(1,10000))
print(sys.getsizeof(L))   #It is used to check the memory size 
print(sys.getsizeof(t))
```

```python
print(list(range(1,10)))
print(tuple(range(1,10)))
```

```python
#Tuples unpacking 

a,b,c=(1,2,3)
print(a)
print(b)
print(c)
```

```python
#Swaping to numbers
a=2
b=3
a,b=b,a
print(a)
print(b)
```

```python
#ZIP in Tuples
t1=(10,4,18)
t2=(2,9,7)
a=tuple(zip(t1,t2))
b=list(zip(t1,t2))
print(a)
print(b)
```

```python
#Sets

#A set is an unordered collection of items. Every set element is unique (no duplicates) 
#and must be immutable (cannot be changed).However, a set itself is mutable. We can add 
#or remove items from it. Sets can also be used to perform mathematical set operations like union, 
#intersection, symmetric difference, etc.

#Characterstics:

#· Unordered
#· Mutable
#· No Duplicates
#. Can't contain mutable data types
```

```python
#Empty set
a={}     #Python treats empty {} as a dictionary by default.
print(type(a))
print(type(set(a)))  #Type conversion is a way to create an empty set
#1D set
a={1,3,5,6,4,1}
print(a)
#2D and 3D sets are not allowed as the sets itself are mutables and mutable datatypes are not allowed in sets.
#Hetrogenous 
a={1,"Hello",3.5,5+6j,True}   #The result won't show the true becuase python treats true as 1 and duplicates are
#not allowed in python.
print(a)
```

```python
s={5,1,"True",1,2,"Work"}
print(s) 
#As we can see the order is not correct, because the sets are unordered collection of items. 
#In sets, the ordering of items happens because of hashing.
```

```python
#Using type conversion
L=[3,1,6,6,4,1]
print(set(L))

#Only immutable datatypes are allowed 
s={5,2,9,(6,6,1,8)} #As tuples are immutable, only tuples are allowed
print(s)
```

```python
#Important concept

s1={1,2,3}
s2={3,1,2}
print(s1==s2)
#The reason is sets are unordered collection of data. In sets order doesn't matter. Hence, == operator
#only comapres the elemets not the order.
```

```python
#Slicing and Indexing
s={4,2,5,7}
print(s[2]) # If I do this it will give me an error, as indexing and slicing are not allowed in sets. Because
#indexing and slicing depend upon index and as sets are inordered. Hence, we don't have any indexing
```

```python
#Adding the items in the sets.
s={1,2,3,4}
s.add(10)
print(s)

#Updating the set 
s1={100,200,300}
s.update(s1)
print(s)
```

```python
s={1,2,3,4}
L=[1,5,8,11,15]
s.update(L)
print(s)
```

```python
#Discard
s={1,200,300,400}
s.discard(300)
print(s)
```

```python
#Clear
s={1,200,300,400}
s.clear()
print(s)
```

```python
#Remove
s={1,200,300,400}
s.remove(200)
print(s)
```

```python
#Pop
s={1,200,300,400}
s.pop()     #pop will remove random number as we don't have a control over it. It is based on Hashing 
#Also, pop in sets does not take any argument.
print(s)
```

```python
#Set operations
#Arithmetic operation (We cannot perform it)
#Comparison Operation (Comparison operators also doesn't perfrom well)
#Bitwise Operators (Doesn't work)
#Logical Operators

s1={1,2,5,6}
s2={1,9,3,10}
s=set()
print(s1 and s2)
print(s1 and set())
print(s1 or set())   #As 1 or 0 is = 1
print(not set())
```

```python
#Membership operators
s1={1,2,5,6}
s2={1,9,3,10}
print(1 in s1)
print(11 in s2)
print(11 not in s1)
```

```python
#Sets in-build Operations

#Union (|)
s1={1,2,3,4}
s2={3,4,5,6}

print(s1.union(s2))
print(s1 | s2)
```

```python
#Intersection (&)
s1={1,2,3,4}
s2={3,4,5,6}

print(s1.intersection(s2))
print(s1 & s2)
```

```python
#Difference (-)
s1={1,2,3,4}
s2={3,4,5,6}

print(s1.difference(s2))
print(s1 - s2)
```

```python
#Difference Update
s1={1,2,3,4}
s2={3,4,5,6}

s1.difference_update(s2)  #It updates the set, by removing the elements found in another set.
print(s1)
```

```python
#Symmetric Difference (^)
s1={1,2,3,4}
s2={3,4,5,6}

print(s1.symmetric_difference(s2))
print(s1 ^ s2)
```

```python
#Symmetric difference update
s1={1,2,3,4}
s2={3,4,5,6}

s1.symmetric_difference_update(s2)
print(s1)
```

```python
#Loop in Sets
s1={1,2,3,4,11,10,"Hello"}
for i in s1:
    print(i)

#Although we can use loops in sets but the order is not in our control
```

```python
#Set functions
#len, min/max, sorted, sum

s1={1,2,3,4}
print(len(s1))
print(sum(s1))
print(min(s1))
print(max(s1))
print(sorted(s1,reverse=True))   #Sorted function does not get affected by hashing problem
```

```python
#issubset
s1={1,2,3,4}
s2={3,4,5,6,1,2}

print(s1.issubset(s2))
```

```python
#issuperset
s1={1,2,3,4}
s2={3,4,5,6}

print(s1.issuperset(s2))
```

```python
#isdisjoint
s1={1,2}
s2={3,4,5,6}

print(s1.isdisjoint(s2))  #Returns True if both the sets has nothing in common
```

```python
#Copy
s1={11,22,33,44}
s2=s1.copy()
print(s2)
```
