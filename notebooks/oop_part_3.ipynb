```python
#Aggregation: Means when one class owns the other class. For example: Customer has a 
#Address. Here the Address class comes under the customer class. Or Smartphone is a class 
#which comes under the product class.

class Customer:
    def __init__(self,name,gender,address):
        self.name=name
        self.gender=gender
        self.address=address

    def __str__(self):
        return "Name-{}, Gender-{}, Address-{}".format(self.name,self.gender,self.address)

    def check_address(self):
        print(f"Address:-{self.address.get_city()}, {self.address.pin}, {self.address.state}")

    def edit_data(self,new_name,new_city,new_pin,new_state):
        self.name=new_name
        self.address.edit_address(new_city,new_pin,new_state)

class Address:
    def __init__(self,city,pin,state):
        self.__city=city
        self.pin=pin
        self.state=state

    def get_city(self):
         return self.__city
    
    def __str__(self):
        return "City-{}, Pin-{}, State-{}".format(self.__city,self.pin,self.state)

    def edit_address(self,new_city,new_pin,new_state):
        self.__city=new_city
        self.pin=new_pin
        self.state=new_state
        
```

```python
add1=Address("LA","110011A","WA")
c1=Customer("John","Male",add1)
print(c1)
```

```python
c1.check_address()
```

```python
c1.edit_data("Lion","SHA","101010A","WWW")   #Using the customer object, changing the
#values in Address Class. Although name changing is happening in Customer class. But the 
#address change is still happening in address class.
```
