```python
#Dictionary

#Dictionary in Python is a collection of keys values, used to store data values like a map, which, 
#unlike other data types which hold only a single value as an element.
#In some languages it is known as map or assosiative arrays.

dict_1 = { 'name' : 'nitish', 'age' : 33 , 'gender' : 'male' }
print(type(dict_1))
```

```python
#Creating an empty dictionary
d={}
print(type(d))

#Creating 1D dictionary
d={1:"Human",2:"animals",3:"Dragons"}
print(d)

#Dictionary with mixed keys
d={(1,2,3):"Hello",1:"Motto"}
print(d)

#2D dictionary
d={1:{"GOD":21,"SUPREME":22},2:"POWER"}
print(d)

s={
    "student":"sup",
    "class":"9th",
    "roll_no.":43,
    "subjcets":{
        "maths":91,
        "chemistry":86,
        "bio":98
    }
}
print(s)
```

```python
#Very Important info
#Creating dictionay using pairs

a=[(1,2),(3,4),(5,6)]
b=((1,2),(3,4),(5,6))
c={(1,2),(3,4),(5,6)}
d=[[1,2],[3,4],[5,6]]
print(dict(a))
print(dict(b))
print(dict(c))
print(dict(d))
```

```python
#Duplicate keys are not allowed
d={"name":"Rahul","name":"sk"}
print(d)
#As we can clearly see Rahul got updated with Sk.
```

```python
#Mutable keys are not allowed
d={[1,2]:"Home",3:"Alone"}
print(d)
```

```python
#Accessing items
#In Dictionaries indexing and slicing is not allowed.
#Dictionaries are ordered datatypes. They keep the sequences just like what we inserted.

d={1:{"GOD":21,"SUPREME":22},2:"POWER"}
print(d[1])
print(d[2])
print(d[1]["SUPREME"])
```

```python
print(d.get("SUPREME"))
#Because supreme is inside a dictionary which is again inside a dictionary. Because of this
#reason we don't use get() instead we do d[1]["SUPREME"]
```

```python
#Adding a key value pair
d={1:{"GOD":21,"SUPREME":22},2:"POWER"}
d[6]="Destruction"
print(d)
```

```python
#Remove key-value pair
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
print(d)
d.pop(3)
print(d)
```

```python
#Clear
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
d.clear()
print(d)
```

```python
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
print(d.popitem())
print(d)
#print(d.popitem(2))  --> Popitem does not take any argument and deletes the last key-value pair
```

```python
#Delete
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
del(d)
print(d)
```

```python
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
del(d[1])   #In this way also we can delete the key-value pairs
print(d)
```

```python
#Deleting 
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
del(d[1]["GOD"])
print(d)
```

```python
#Editing a dictionary
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
d[1]["GOD"]=40
print(d)
```

```python
#Dictionary Operations
#No Arthimetic Operations are allowed
#Comparison Operators not allowed
#Bitwise Operators

#Logical Operators
d1={"Hello":1,"World":2}
d2={"Hello":1,"world":3}
d={}

print(d1 and d2)
print(d1 or d)
print(d1 and d)
print(not d1)
```

```python
#Membership operators
d1={"Hello":1,"World":2}

print(1 in d1)       #False because in dictionaries everything is about keys. And 1 is not a key
print("Hello" in d1)
```

```python
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
for i in d:
    print(i)

#Again, as we can see only keys are getting printed because in dictionaries, keys are more important.
```

```python
#Printing key value pairs

for i,j in d.items():
    print(i,j)
```

```python
for i in d:
    print(i,d[i])
```

```python
#Dictionaries functions
#Len
#min/max
#sorted

d={2:{"GOD":21,"SUPREME":22},1:"POWER",4:"To be x"}
print(len(d))
print(min(d))
print(max(d))
print(sorted(d))  #Giving us the sorted sequence of keys. 
```

```python
#items, keys, values
d={1:{"GOD":21,"SUPREME":22},2:"POWER",3:"To be x"}
print(d.items())
print(d.keys())
print(d.values())
```

```python
d1={"Hell":12,"World":2}
d2={"Hello":1,"world":3}

d1.update(d2)
print(d1)
```

```python
#Dictionary Comprehension
#{keys:value for vars in iterable}
d={1:2,3:4,5:6}

d={i:j**2 for i,j in d.items()}
print(d)
```

```python
#printing the first 10 numbers and their squares
d={i:i**2 for i in range(1,10)}
print(d)
```

```python
## using existing dict convert the distance from km to miles
distances = {'delhi' :1000, 'mumbai' :2000, 'bangalore':3000}

d={i:j*0.62 for i,j in distances.items()}
print(d)
```

```python
## using zip
days=["Sunday", "Monday", "Tuesday", "Wednesday","Thursday", "Friday", "Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

a=dict(zip(days,temp_C))
print(a)
```

```python
# using if condition to show only those products, that are in stock
products = {'phone' : 10, 'laptop' : 0,'charger' : 32, 'tablet' : 0}

d={i:products[i] for i in products if products[i]>0}
print(d)
```

```python
#A really good example
# Nested Comprehension
# print tables of number from 2 to 4

# {
#2: {1:2, 2:4, 3:6,4:8},
#3: {1:3, 2:6, 3:9,4:12},
#4: {1:4, 2:8, 3:12,4:16}
# }

d={
    j:{i:i*j 
    for i in range(1,11)}
    for j in range(2,5)
}
print(d)
```
