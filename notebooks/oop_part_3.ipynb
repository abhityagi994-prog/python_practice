```python
#Aggregation: Means when one class owns the other class. For example: Customer has a 
#Address. Here the Address class comes under the customer class. Or Smartphone is a class 
#which comes under the product class.

class Customer:
    def __init__(self,name,gender,address):
        self.name=name
        self.gender=gender
        self.address=address

    def __str__(self):
        return "Name-{}, Gender-{}, Address-{}".format(self.name,self.gender,self.address)

    def check_address(self):
        print(f"Address:-{self.address.get_city()}, {self.address.pin}, {self.address.state}")

    def edit_data(self,new_name,new_city,new_pin,new_state):
        self.name=new_name
        self.address.edit_address(new_city,new_pin,new_state)

class Address:
    def __init__(self,city,pin,state):
        self.__city=city
        self.pin=pin
        self.state=state

    def get_city(self):
         return self.__city
    
    def __str__(self):
        return "City-{}, Pin-{}, State-{}".format(self.__city,self.pin,self.state)

    def edit_address(self,new_city,new_pin,new_state):
        self.__city=new_city
        self.pin=new_pin
        self.state=new_state
        
```

```python
add1=Address("LA","110011A","WA")
c1=Customer("John","Male",add1)
print(c1)
```

```python
c1.check_address()
```

```python
c1.edit_data("Lion","SHA","101010A","WWW")   #Using the customer object, changing the
#values in Address Class. Although name changing is happening in Customer class. But the 
#address change is still happening in address class.
```

```python
c1.check_address()
```

```python
class Parent:
    def __init__(self,num):
        self.__num=num

    def get_num(self):
        return self.__num

class Child(Parent):
    def __init__(self,val,num):
        self.__val=val

    def get_val(self):
        return self.__val
```

```python
C=Child(100,10)
```

```python
print(C.get_num())  #This will throw an error because, get_num is the function, which 
#is printing the value of private member __num. But as the parent class constructor 
#never executes, __num never receives the  value.
```

```python
#Method Overriding 

class Phone:
    def __init__(self,price,brand,camera):
        print("I am inside the Phone class")
        self.__price=price
        self.brand=brand
        self.camera=camera

    def buy(self):
        print("I am buy of Phone Class")

class SmartPhone(Phone):
    def buy(self):
        print("I am buy of SmartPhone Class")
```

```python
a=SmartPhone(100000,"Samsung","48Mpix")
```

```python
a.buy()    #This is called Method Overriding. As we have same buy() functions in both
#the classes but the only buy() of child class will get execute, even though it does
#not have it's own constructor.
```

```python
#SuperKey
class Phone:
    def __init__(self,price,brand,camera):
        self.price=price
        self.brand=brand
        self.camera=camera

    def buy(self):
        print("I am buy of Phone Class")

class SmartPhone(Phone):
    def buy(self):
        print("I am buy of Child Class")
        super().buy()
```

```python
#SuperKey anothe example
class Phone:
    def __init__(self,price,brand,camera):
        print("I am __init__ of Phone")
        self.price=price
        self.brand=brand
        self.camera=camera

    def buy(self):
        print("I am buy of Phone Class")

class SmartPhone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        print("I am __init__ of SmartPhone")
        super().__init__(price,brand,camera)
        self.os=os
        self.ram=ram
        print("I am __init__ of SmartPhone")

    def buy(self):
        print("I am buy of SmartPhone class")
```

```python
#Method resolution order

class Phone:
    def __init__(self,price,brand,camera):
        self.price=price
        self.brand=brand
        self.camera=camera

    def buy(self):
        print("I am buy of Phone class")

class Product:
    def buy(self):
        print("I am buy of Product class")

class SmartPhone(Phone,Product):
    pass


a=SmartPhone(8888,"Gione","33Mpix")
a.buy()
#As we can see buy method of Phone class gets printed. Because in SmartPhone class we have written Phone, Product
#when Phone is first. That's why it the method of Phone class was executed. But if we do like this -
#SmartPhone(Product,Phone), then buy() method of Product class will get executed. And this is known as Method
#Resolution order (MRO)
```

```python
#Polymorphism means one interface and multiple forms. For ex - Len() function, which 
#behave differently for different datatypes, but calculate the same thing: Length.


class shape:

    def area(self,radius):
        return 3.14*(radius)**2

    def area(self,length,breadth):
        return length * breadth
```

```python
a=shape()
```

```python
print(a.area(4)) #But this will throw an error like missing one argument. As, even if we
#are passing one argument, python is not considering the circus area. It is only looking 
#at the latest area function which is of rectangle.
#This is waht we call method overloading. 
```

```python
#Solving the problem of methods overloading - 

class shape:

    def area(self,a,b=0):
        if b==0:
            return 3.14*(a)**2
        else:
            return a*b
            
```

```python
a=shape()
```

```python
print(a.area(4))
print(a.area(4,5))
```

```python
#Abstraction means showing only what is necessary and hiding the unnecessary details.

from abc import ABC,abstractmethod

class BankApp(ABC):
    def __init__(self):
        self.database={}

    @abstractmethod
    def security(self):
        pass

    @abstractmethod
    def display(self):
        pass

    def tota_amount(self):
        print("No cash found")

class MobileApp(BankApp):
    def security(self):
        print("Securrity Added")

    def display(self):
        print("Display added")
```
