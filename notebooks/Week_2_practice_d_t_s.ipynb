```python
#Q1: Join Tuples if similar initial element

#While working with Python tuples, we can have a problem in which we need to perform concatenation of records from 
#the similarity of initial element. This problem can have applications in data domains such as Data Science.

#For eg.

#Input : 
test_list = [(5,6),(5,7),(5,8),(6,10),(7,13)]
#Output : [(5,6,7,8),(6,10),(7,13)]

unique=[]
for i in test_list:
    unique.append(i[0])

unique=set(unique)
print(unique)

result=[]
for i in unique:
    result.append([i])
    for j in test_list:
        if i==j[0]:
            result[-1].append(j[1])
print(result)

print(list(map(tuple,result)))
```

```python
#Q2: Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.

#For eg.
#The original tuple : (1, 5, 7, 8, 10)
t=(1,5,7,8,10)
#Resultant tuple after multiplication :

#(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

#output-(5, 40, 91, 136, 80)
L=[]
for i in range(0,len(t)):
    if i==0:
        L.append(t[i]*t[i+1])
    elif i==(len(t)-1):
        L.append(t[i]*t[i-1])
    else:
        L.append(t[i]*t[i-1]+t[i]*t[i+1])

print(L)
```

```python
#Q3 : Check is tuples are same or not?

#Two tuples would be same if both tuples have same element at same index

t1 = (1,2,3,0)
t2 = (1,2,0,3)

#t1 and t2 are not same

flag=True
for i,j in tuple(zip(t1,t2)):
    if i!=j:
        flag=False
        break
if flag:
    print("Same Tuple")
else:
    print("Different Tuple")
```

```python
#Q4 : Count no of tuples, list and set from a list

list1 = [{'hi', 'bye'}, {'Geeks', 'forGeeks'}, ('a', 'b'), ['hi', 'bye'], ['a', 'b']]

#Output:
#List-2
#Set-2
#Tuples-1

d={"List":0,"Set":0,"Tuples":0}
for i in list1:
    if type(i)==list:
        d["List"]=d["List"]+1
    elif type(i)==set:
        d["Set"]=d["Set"]+1
    elif type(i)==tuple:
        d["Tuples"]=d["Tuples"]+1

for i,j in d.items():
    print(i,"-",j)
        
```

```python
#Q2: Write a program to count unique number of vowels using sets in a given string. Lowercase and upercase
#vowels will be taken as different.

#Input:

Str1 ="hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"

#Output:
#No of unique vowels-6

L=[]

for i in Str1.split():
    for j in i:
        if j in ("a","A","e","E","i","I","o","O","u","U"):
            L.append(j)
            
b=set(L)
print("No. of unique vowels",len(b))
```

```python
#Q4: find union of n arrays.

#Example 1:
#Input:
L=[[1, 2, 2, 4, 3, 6],
[5, 3, 4],
[9, 5, 7, 1],
[2, 4, 1, 3]]

#Output: 
#[1,2,3,4,5,6,7,9]

S=set()
for i in L:
    for j in i:
        S.add(j)

print(list(S))
print("\n")
S1=set()
for i in L:
    S1.update(i)

print(list(S1))
```

```python
#Q5 : Intersection of two lists. Intersection of two list means we need to take all those 
#elements which are common to both of the initial lists and store them into another list. 
#Only use using list-comprehension.

#Example 1:
#Input:

lst1 = {15, 9, 10, 56, 23, 78, 5, 4, 9}
lst2 = {9, 4, 5, 36, 47, 26, 10, 45, 87}

#Output:
#[9, 10, 4, 5]
#S={}
S=[i for i in lst1 if i in lst2]
print(S)
```

```python
#Dictionary

#Q1: Key with maximum unique values
#Given a dictionary with values list, extract key whose value has most unique values.

#Example 1:
#Input:
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "Best" : [9, 9, 6, 5, 5]}

#Output:
#CampusX
key=""
max_key_length=0
for i,j in test_dict.items():
    if len(set(j))>max_key_length:
        max_key_length=len(set(j))
        key=i

print(key)
```

```python
#Q2: Replace words from Dictionary. Given String, replace it's words from lookup ictionary.

#Example 1:
#Input:
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}

#Output:
#CampusX is the best channel for Data-Science students.

final=[]
for i in test_str.split():
    if i in repl_dict:
        final.append(repl_dict[i])
    else:
        final.append(i)

final_=" ".join(final)
print(final_)
```

```python
#Q3 : Convert List to List of dictionaries. Given list values and keys list, convert 
#these values to key value pairs in form of list of dictionaries.

#Example 1:

#Input:
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]

#Output:
#[{'name': 'DataScience', 'id': 3}, {'name' : 'is', 'id': 8}]

result=[]
for i in range(0,len(test_list),2):
    result.append({key_list[0]:test_list[i],key_list[1]:test_list[i+1]})

print(result)
```
