```python
#Nested if with List Comprehension
basket = ['apple' , 'guava' ,'cherry' , 'banana' ]
my_fruits = ['apple' , 'kiwi' , 'grapes' , 'banana' ]

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'
L_fruits=[i
          for i in my_fruits
          for j in basket
          if i in j and i[0]=="a"
         ]
print(L_fruits)
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
#A really good example

L=[[i*j 
   for i in range(1,4)]  #Doing this we are creating a list inside a list.
   for j in range(1,4)
  ]
print(L)
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

L=[[i*j
   for i in L2]
   for j in L1
  ]

print(L)
```

```python
#Zip
#The zip() function returns a zip object, which is an iterator of tuples where the first item in each 
#passed iterator is paired together, and then the second item in each passed iterator are paired together. 
#If the passed iterators have different lengths, the iterator with the least items decides 
#the length of the new iterator.

# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

L3=[L1[i]+L2[i] for i in range(0,len(L1))]
print(L3)
```

```python
#Now solving the above question using the zip
a=list(zip(L1,L2))
print(a)

L3=[i+j for i,j in a]
print(L3)
```

```python
# Create 2 lists from a given list where
# 1st list will contain all the odd numbers from the original list and
# the 2nd one will contain all the even numbers

L = [1,2,3,4,5,6]
L_odd=[]
L_even=[]
for i in L:
    if i%2==0:
        L_even.append(i)
    else:
        L_odd.append(i)

print(L_even)
print(L_odd)
```
