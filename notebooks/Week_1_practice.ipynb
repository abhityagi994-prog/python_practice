```python
#Problem 8: Take a user input as integer N. Find out the sum from 1 to N. If any number
#if divisible by 5, then skip that number. And if the sum is greater than 300, don't need
#to calculate the sum further more.
#Print the final result. And don't use for loop to solve this problem.

nth=int(input("Please enter the nth value: "))
s=0
i=0
while i<=nth:
    i+=1
    if i%5!=0:
        if s+i<300:
            s=s+i
        else:
            break

print(f"{i} and {s}")
```

```python
#Reverse a given integer number
#For example:
#Input - 76542
#Output - 24567
rev=0
a=int(input("Enter the integer num: "))
while a!=0:
    rev=rev*10+a%10
    a=a//10

print(rev)
print(type(rev))
```

```python
#Write a program to count the number of words in a string without split()

string=input("Please enter the string: ")
L=[]
temp=""
for i in string:
    if i!=" ":
        temp=temp+i
    else:
        L.append(temp)
        temp=""

L.append(temp)
print(len(L))
```

```python
#Write a program to check whether a given string is Palindrome or not

user_input=input("Enter the string: ")
while True:
    if len(user_input)==1 or len(user_input)==0:
        print("Palindrome")
        break
    else:
        if user_input[0]==user_input[-1]:
            user_input=user_input[1:-1]
        else:
            print("Not a Palindrome")
            break
```
